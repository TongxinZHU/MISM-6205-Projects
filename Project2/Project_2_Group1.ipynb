# Conduct Data Cleansing

```python
#Drop the Class column
df_Business.drop(['Class'], axis = 1, inplace = False)
df_Business = df_Business.reset_index(drop = True)
df_Business.head()
```

```python
df_Business.columns
```

```python
#Rename columns
df_Business.rename(columns = lambda x: x.lower().replace(' ','_'), inplace= True)
df_Business.columns
```

### Modify improperly formatted data and handle bad data
Base on the dsta dictionary provided

```python
#Fix the gender column
#Replace 'F', 'M' with 'Female', 'Male'
df_Business['gender'] = df_Business['gender'].replace(['F', 'M'],['Female', 'Male'])
```

```python
#Check updated gender column
df_Business['gender'].unique()
```

```python
#Fix the customer_type column
#Replace 'disloyal' with 'Disloyal Customer' and 'loyal', 'Loyal' with 'Loyal Customer'
df_Business['customer_type'] = df_Business['customer_type'].replace(['disloyal', 'loyal', 'Loyal'],['Disloyal Customer', 'Loyal Customer', 'Loyal Customer'])
```

```python
#Check updated customer_type column
df_Business['customer_type'].unique()
```

```python
#Fix the age column
#Replace 180 with 80, and 185 with 85 
#Replace 0 with the first mode value 
df_Business['age'] = df_Business['age'].replace([180, 185],[80, 85])
mode_value = df_Business['age'].mode()
df_Business['age'] = df_Business['age'].replace([0],[mode_value[0]])
```

```python
#Check updated age column
df_Business['age'].unique()
```

```python
#Fix the type_of_travel column 
#Replace 'Business', 'Business travel' with "Business Travel"
df_Business['type_of_travel'] = df_Business['type_of_travel'].replace(['Business', 'Business travel'],['Business Travel', 'Business Travel'])
```

```python
#Check updated type_of_travel column
df_Business['type_of_travel'].unique()
```

```python
#Fix the flight_distance column 
#Replace negative value with a positive one
#Replace 0 with the first mode value
df_Business['flight_distance'] = df_Business['flight_distance'].replace([-5.421e+03],[5.421e+03])
mode_value = df_Business['flight_distance'].mode()
df_Business['flight_distance'] = df_Business['flight_distance'].replace([0.000e+00],[mode_value[0]])
```

```python
#Check updated flight_distance column
df_Business['flight_distance'].unique()
sorted(df_Business['flight_distance'].unique())
```

```python
#Check unique values of inflight_wifi_service column
df_Business['inflight_wifi_service'].unique()
```

```python
#Fix the departure/arrival_time_convenient column
#Replace '0' with '1' becase we assume 0 indicates 'not good'
df_Business['departure/arrival_time_convenient'] = df_Business['departure/arrival_time_convenient'].replace([0],[1])
```

```python
#Check updated departure/arrival_time_convenient column
df_Business['departure/arrival_time_convenient'].unique()
```

```python
#Fix the ease_of_online_booking column
#Replace '0' with '1' and 'Unknown' with the first mode value
mode_value = df_Business['ease_of_online_booking'].mode()
df_Business['ease_of_online_booking'] = df_Business['ease_of_online_booking'].replace(['0', 'Unknown'],['1', mode_value[0]])
```

```python
#Check updated ease_of_online_booking column
df_Business['ease_of_online_booking'].unique()
```

```python
#Fix the food_and_drink column
#Replace 'Great' with '5' because 5 is the highest score which can be identified as great, and '0' with '1'
#Also, to unify the data format, replace int 5 with string '5' because there are still some null values in the column
#We will convert the data type later
mode_value = df_Business['food_and_drink'].mode()
df_Business['food_and_drink'] = df_Business['food_and_drink'].replace(['Great', 5, '0'],['5', '5', '1'])
```

```python
#Check updated food_and_drink' column
df_Business['food_and_drink'].unique()
```

```python
#Fix the inflight_entertainment column
#Replace '0' with '1', and '10' with '5' because 10 is a high score
#Some poeple might think the highest score is 10
mode_value = df_Business['inflight_entertainment'].mode()
df_Business['inflight_entertainment'] = df_Business['inflight_entertainment'].replace([0, 10],[1, 5])
```

```python
#Check updated inflight_entertainment column
df_Business['inflight_entertainment'].unique()
```

```python
#Fix the baggage_handling column
#Replace 'A' with '5' since A is always idtified as a good score, and unify the data format. 
df_Business['baggage_handling'] = df_Business['baggage_handling'].replace(['A', 5],['5', '5'])
```

```python
#Check updated baggage_handling column
df_Business['baggage_handling'].unique()
```

```python
#Fix the cleanliness column
#Replace '0.' with '1'
df_Business['cleanliness'] = df_Business['cleanliness'].replace([0.],[1])
```

```python
#Check updated cleanliness column
df_Business['cleanliness'].unique()
```

```python
#Fix the departure_delay_in_minutes column
#Replace the negative value with a positive one
df_Business['departure_delay_in_minutes'] = df_Business['departure_delay_in_minutes'].replace([-2.000e+00], [2.000e+00])
```

```python
#Check updated departure_delay_in_minutes column
sorted(df_Business['departure_delay_in_minutes'].unique())
```

```python
#Fix the arrival_delay_in_minutes column
#Replace the negative value with a positive one
df_Business['arrival_delay_in_minutes'] = df_Business['arrival_delay_in_minutes'].replace([-4.500e+01], [4.500e+01])
```

```python
#Check updated arrival_delay_in_minutes column
sorted(df_Business['arrival_delay_in_minutes'].unique())
```

```python
#Fix the satisfaction column
#Replkace 'satisfied' with 'Satisfied', 'dissatisfied' with 'Dissatisfied', and 'neutral' with 'Neutral'
df_Business['satisfaction'] = df_Business['satisfaction'].replace(['satisfied', 'dissatisfied', 'neutral'], ['Satisfied', 'Dissatisfied', 'Neutral'])
```

```python
#Check updated satisfaction column
df_Business['satisfaction'].unique()
```

```python
#Check updated dataset
df_Business
```

```python
#Missing values
df_Business.isnull().sum()
```

```python
#Fill missing values of categorical columns with 'Unknown' and numerical columns with the first mode value of each column
df_Business['gender'].fillna('Unknown', inplace = True)
df_Business['customer_type'].fillna('Unknown', inplace = True)
df_Business['age'].fillna(df_Business['age'].mode()[0], inplace = True)
df_Business['type_of_travel'].fillna('Unknown', inplace = True)
df_Business['flight_distance'].fillna(df_Business['flight_distance'].mode()[0], inplace = True)
df_Business['inflight_wifi_service'].fillna(df_Business['inflight_wifi_service'].mode()[0], inplace = True)
df_Business['departure/arrival_time_convenient'].fillna(df_Business['departure/arrival_time_convenient'].mode()[0], inplace = True)
df_Business['ease_of_online_booking'].fillna(df_Business['ease_of_online_booking'].mode()[0], inplace = True)
df_Business['food_and_drink'].fillna(df_Business['food_and_drink'].mode()[0], inplace = True)
df_Business['inflight_entertainment'].fillna(df_Business['inflight_entertainment'].mode()[0], inplace = True)
df_Business['baggage_handling'].fillna(df_Business['baggage_handling'].mode()[0], inplace = True)
df_Business['cleanliness'].fillna(df_Business['cleanliness'].mode()[0], inplace = True)
df_Business['departure_delay_in_minutes'].fillna(df_Business['departure_delay_in_minutes'].mode()[0], inplace = True)
df_Business['arrival_delay_in_minutes'].fillna(df_Business['arrival_delay_in_minutes'].mode()[0], inplace = True)
df_Business['satisfaction'].fillna('Unknown', inplace = True)
```

```python
#Check if there are still missing values
df_Business.isnull().sum()
```

```python
#Remove the fully duplicated observations
df_Business.drop_duplicates(inplace = True)
```

```python
#Check if there are still duplicated rows
df_Business.duplicated().sum()
```

```python
#Fix the columns datatypes
#Covert 'gender', 'customer_type', 'type_of_travel', 'class', 'satisfaction' to string
#Covert 'flight_distance', 'departure_delay_in_minutes', 'arrival_delay_in_minutes' to float
#Covert rest columns to int
df_Business['age'] = df_Business['age'].astype("Int64")
df_Business['inflight_wifi_service'] = df_Business['inflight_wifi_service'].astype("Int64")
df_Business['departure/arrival_time_convenient'] = df_Business['departure/arrival_time_convenient'].astype("Int64")
df_Business['ease_of_online_booking'] = df_Business['ease_of_online_booking'].astype('float').astype('Int64')
df_Business['food_and_drink'] = df_Business['food_and_drink'].astype('float').astype('Int64')
df_Business['inflight_entertainment'] = df_Business['inflight_entertainment'].astype('float').astype('Int64')
df_Business['baggage_handling'] = df_Business['baggage_handling'].astype('float').astype('Int64')
df_Business['cleanliness'] = df_Business['cleanliness'].astype('float').astype('Int64')
df_Business['gender'] = df_Business['gender'].astype("string")
df_Business['customer_type'] = df_Business['customer_type'].astype("string")
df_Business['type_of_travel'] = df_Business['type_of_travel'].astype("string")
df_Business['class'] = df_Business['class'].astype("string")
df_Business['satisfaction'] = df_Business['satisfaction'].astype("string")
```

```python
#Updated data types
df_Business.info()
```

```python
#Final version of dataset
df_Business
```
